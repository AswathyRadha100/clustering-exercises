# Visualize the clusters of model1

```python
plt.figure(figsize=(8, 5))

for cluster_n in train.model1_cluster_predict.sort_values().unique():
    plt.scatter(train[train.model1_cluster_predict == cluster_n].spending_score, train[train.model1_cluster_predict == cluster_n].annual_income, label=f'cluster_{cluster_n}')

plt.legend()
plt.title("Model1 K-Means Clusters")
plt.show()
```

```python
# Visualizing the change in inertia as the number of clusters (k) increases.

with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(9, 6))
    pd.Series({k: KMeans(k).fit(X).inertia_ for k in range(2, 12)}).plot(marker='x')
    plt.xticks(range(2, 12))
    plt.xlabel('k')
    plt.ylabel('inertia')
    plt.title('Change in inertia as k increases')
```

# Use the elbow method and select  visually  a good value for k


### Lets select K= 6

```python
# Define custom colors for the clusters
cluster_colors = ['blue', 'orange','yellow','green','red','m']  

# Create a colormap with the custom colors
cmap = mcolors.ListedColormap(cluster_colors)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(13, 13), sharex=True, sharey=True)

for ax, k in zip(axs.ravel(), range(3, 7)):
    clusters = KMeans(k).fit(X).predict(X)
    ax.scatter(train.spending_score, train.annual_income, c=clusters, cmap=cmap)  
    ax.set(title='k = {}'.format(k), xlabel='age', ylabel='annual_income')

plt.show()
```
